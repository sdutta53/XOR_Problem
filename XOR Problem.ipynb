```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
from tensorflow import keras
```

```python
x_data = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

y_data = [
    [0],
    [1],
    [1],
    [0]
]
```

```python
x_data=np.array(x_data)
y_data=np.array(y_data)
```

```python
model = keras.Sequential()
```

```python
model.add(keras.layers.Dense(15,activation="sigmoid",input_shape=(2,)))
model.add(keras.layers.Dense(1,activation="sigmoid"))
```

```python
optimizer= keras.optimizers.SGD(lr=0.5)
model.compile(optimizer=optimizer, loss="binary_crossentropy",metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_data,y_data,batch_size=4,epochs=2000)
```

```python
predict=model.predict(x_data)
print(predict)
```

```python
x_samples=[[0,0]]
for i in range(101):
    for j in range(101):
        x_samples.append([i/100,j/100])
x_samples=np.array(x_samples)
```

```python
results=model.predict(x_samples)
```

```python
temp=np.array_split(x_samples,2,axis=1)
X=temp[0]
Y=temp[1]
Z=results
#print(X,Y)
```

```python
%matplotlib notebook
fig  = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.scatter(X,Y,Z, c='r', marker='.')
ax.set_title('XOR Gate')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Prediction')

plt.show()
```
